```python
import pandas as pd
import piplite
await piplite.install("seaborn")
import pyodide
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

## 1 出生率のデータ

```python
data = pd.read_csv(pyodide.open_url("https://raw.githubusercontent.com/ChungWookyung/kg-jupyterlite-data-analysis/main/content/DATA/birth.csv"))
```

```python
data.head()
```

日本の各県の出生率(千人あたり)データ

```python
data["Date"].unique()
```

```python
data.groupby("Ken").describe()["Birth"]
```

```python
tokyodata = data.loc[data["Ken"] == "tokyo",:]
tokyodata.head()
```

```python
sns.lineplot(data=tokyodata, x="Date", y="Birth")
```

```python
kenlist = ["osaka","tokyo","hokkaido"]
data2 = data.loc[data["Ken"].isin(kenlist),:]
data2.head()
```

```python
sns.lineplot(data = data2, x="Date", y="Birth", hue="Ken")
```

## 3 サービス売上

```python
data = pd.read_csv(pyodide.open_url("https://raw.githubusercontent.com/ChungWookyung/kg-jupyterlite-data-analysis/main/content/DATA/keizai.csv"))
```

```python
data.head()
```

サービス産業の売上データ
- Service: サービス産業全体
- Logistic: 物流
- House: 不動産
- Tech: 技術
- Food: 宿泊・飲食
- Enter: 娯楽
- Edu: 教育
- Health: 医療

```python
data["Date"].unique()
```

```python
data.groupby(["Ken"]).describe()["Service"]
```

```python
data2017 = data.loc[data["Date"]==2017,:]
data2017.head()
```

```python
sns.barplot(data = data2017, y="Ken", x="Service")
```

```python
sns.barplot(data = data2017, y="Ken", x="Tech")
```

```python
vlist = data.columns[3:]
tokyodata = data.loc[data["Ken"] == "tokyo",:]
tokyodata = tokyodata.melt(id_vars="Date",value_vars=vlist)
tokyodata.head()
```

```python
sns.lineplot(data = tokyodata, x = "Date", y="value", hue="variable")
```
