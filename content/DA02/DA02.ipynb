## 2.3 条件文の操作

```python
dt.head()
```

```python
dt[dt["国語"] > 80]
```

```python
dt[dt["性別"] == "女"]
```

```python
dt[dt["学年"] == 3]
```

**実習：条件文で以下を取り出してください**

1. "数学" が90以上の生徒
2. "身長" が170未満の生徒
3. "クラス" が "B" の生徒

## 2.4 条件文の重複

```python
dt[(dt["性別"] == "女") & (dt["学年"] == 3)]
```

```python
dt[(dt["国語"] >= 90) | (dt["数学"] >= 90)]
```

**実習：条件文の重複で以下を取り出してください**

1. "成績合計" が170以上で **かつ** "クラス" が "A" の生徒
2. "身長" が170以上、**もしくは** "体重" が70以上の生徒

## 4.2 国別パネルデータ（US/Japan/China, 2014–2023）

```python
panel = pd.read_csv(pyodide.http.open_url("https://raw.githubusercontent.com/ChungWookyung/kg-jupyterlite-data-analysis/main/content/DATA/country_panel.csv"))
panel.head()
```

```python
panel.groupby("country").mean(numeric_only=True)
```

```python
panel.groupby("country")["gdp_trillion_usd"].mean()
```

```python
panel.groupby("year")["gdp_trillion_usd"].sum()
```

```python
panel.groupby("country").count()
```

```python
panel.groupby("country")["gdp_trillion_usd"].agg(["mean", "max", "min"])
```

**実習：パネルデータと groupby で以下を計算してください**

1. 国ごとの平均人口
2. 年ごとの3か国合計人口
3. `agg(["mean","std"])` を使って、国ごとのGDPの平均と標準偏差を同時に出す
